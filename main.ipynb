# fill the null values with the median of that particular category

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns ={'purchase_amount_(usd)':'purchase_amount'})
df = df.rename(columns ={'frequency_of_purchases\r':'frequency_of_purchases'})
```

```python
df.columns
```

```python
labels = ['Young_adult','adult','middle_aged','senior']
df['age_group'] = pd.qcut(df['age'],q =4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase frequency days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['frequency_of_purchases'] = df['frequency_of_purchases'].str.replace('\r', '', regex=False)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
